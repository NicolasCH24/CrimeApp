```python
def extract_data(dict_csvs):
    def create_df_year(dict_csvs, año):
        dataframes_meses = []
        df_año = pd.DataFrame()
        for mes in ['enero', 'febrero','marzo','abril','mayo','junio','julio','agosto','septiembre','octubre','noviembre','diciembre']:
            df = pd.read_csv(dict_csvs[año][0][mes])
            dataframes_meses.append(df)

        for df1 in dataframes_meses:
            df_año = pd.concat([df_año, df1])
        
        return df_año
    
    df_2016 = create_df_year(dict_csvs, 2016)
    df_2017 = create_df_year(dict_csvs, 2017)
    df_2018 = create_df_year(dict_csvs, 2018)
    df_2019 = create_df_year(dict_csvs, 2019)
    df_2020 = create_df_year(dict_csvs, 2020)
    df_2021 = create_df_year(dict_csvs, 2021)
    df_2022 = create_df_year(dict_csvs, 2022)
    df_2023 = create_df_year(dict_csvs, 2023)

    dataframes  = [df_2016, df_2017, df_2018, df_2019, df_2020, df_2021, df_2022, df_2023]

    def delitos_unicos(dataframes):
        delitos_comunes = set(dataframes[0]['Tipo'].unique())

        for df in dataframes[1:]:
            delitos_comunes.intersection_update(df['Tipo'].unique())

        return delitos_comunes

    lista_delitos_comunes = delitos_unicos(dataframes)

    df = pd.concat([df_2016, df_2017, df_2018, df_2019, df_2020, df_2021, df_2022, df_2023])
    df = df[df['Tipo'].isin(lista_delitos_comunes)]
    df_entrada = df.dropna()

    return df_entrada
    
```

```python
df_entrada = extract_data(dict_csvs)
```

```python
def transform_fact_table(df_entrada):
    # CONTACTO_ID
    df_entrada = df_entrada.reset_index()
    df_entrada = df_entrada.rename(columns={'index':'CONTACTO_ID'})

    # TIPO_DELITO_KEY
    def transform_tipo(tipo):
        if tipo == 'Robo':
            return 1
        elif tipo == 'Robo automotor':
            return 2
        elif tipo == 'Hurto':
            return 3
        elif tipo == 'Hurto automotor':
            return 4
        elif tipo == 'Amenazas':
            return 5
        elif tipo == 'Lesiones Dolosas':
            return 6
        elif tipo == 'Homicidios dolosos':
            return 7
        elif tipo == 'Lesiones en siniestros viales':
            return 8
        else:
            return "Delito no encontrado"
        
    df_entrada['TIPO_DELITO_KEY'] = df_entrada['Tipo'].apply(transform_tipo)

    # BARRIO_KEY
    barrio_dict = {
    'RETIRO': 1, 'SAN NICOLAS': 2, 'PUERTO MADERO': 3, 'SAN TELMO': 4, 'MONSERRAT': 5, 'CONSTITUCION': 6,
    'RECOLETA': 7, 'BALVANERA': 8, 'SAN CRISTOBAL': 9, 'BOCA': 10, 'BARRACAS': 11, 'PARQUE PATRICIOS': 12, 
    'NUEVA POMPEYA': 13, 'ALMAGRO': 14, 'BOEDO': 15, 'CABALLITO': 16, 'FLORES': 17, 'PARQUE CHACABUCO': 18,
    'VILLA SOLDATI': 19, 'VILLA RIACHUELO': 20, 'VILLA LUGANO': 21, 'LINIERS': 22, 'MATADEROS': 23, 
    'PARQUE AVELLANEDA': 24, 'VILLA REAL': 25, 'MONTE CASTRO': 26, 'VERSALLES': 27, 'FLORESTA': 28, 
    'VELEZ SARSFIELD': 29, 'VILLA LURO': 30, 'VILLA DEVOTO': 31, 'VILLA DEL PARQUE': 32, 'VILLA SANTA RITA': 33, 
    'VILLA GRAL. MITRE': 34, 'COGHLAN': 35, 'SAAVEDRA': 36, 'VILLA URQUIZA': 37, 'VILLA PUEYRREDON': 38, 
    'NUÑEZ': 39, 'BELGRANO': 40, 'COLEGIALES': 41, 'PALERMO': 42, 'CHACARITA': 43, 'VILLA CRESPO': 44, 
    'PATERNAL': 45, 'VILLA ORTUZAR': 46, 'AGRONOMIA': 47, 'PARQUE CHAS': 48
    }

    df_entrada['BARRIO_KEY'] = df_entrada['Barrio'].map(barrio_dict)

    # COMUNA_KEY
    def transform_comuna(comuna):
        if comuna == 'COMUNA 1':
            return 1
        elif comuna == 'COMUNA 2':
            return 2
        elif comuna == 'COMUNA 3':
            return 3
        elif comuna == 'COMUNA 4':
            return 4
        elif comuna == 'COMUNA 5':
            return 5
        elif comuna == 'COMUNA 6':
            return 6
        elif comuna == 'COMUNA 7':
            return 7
        elif comuna == 'COMUNA 8':
            return 8
        elif comuna == 'COMUNA 9':
            return 9
        elif comuna == 'COMUNA 10':
            return 10
        elif comuna == 'COMUNA 11':
            return 11
        elif comuna == 'COMUNA 12':
            return 12
        elif comuna == 'COMUNA 13':
            return 13
        elif comuna == 'COMUNA 14':
            return 14
        elif comuna == 'COMUNA 15':
            return 15
        
    df_entrada['COMUNA_KEY'] = df_entrada['Comuna'].apply(transform_comuna)

    fct_hechos = df_entrada[['CONTACTO_ID', 'Fecha', 'Franja Horaria', 'TIPO_DELITO_KEY', 'Latitud', 'Longitud', 'BARRIO_KEY', 'COMUNA_KEY']]
    fct_hechos = fct_hechos.rename(columns={'Fecha':'FECHA', 'Franja Horaria':'FRANJA_HORARIA', 'Latitud':'LATITUD', 'Longitud':'LONGITUD'})

    return fct_hechos
```
